```python
gender = pd.get_dummies(df['Gender'],drop_first=True)
married = pd.get_dummies(df['Married'],drop_first=True)
Property = pd.get_dummies(df['Property_Area'],drop_first=False)
df.drop(['Gender',"Married","Property_Area"],axis=1,inplace=True)
df = pd.concat([df,gender,married,Property],axis=1)
```

```python
df.rename(columns={"Yes":"Married"} ,inplace=True)
```

```python
self_employed = pd.get_dummies(df['Self_Employed'],drop_first=True)
loan_status = pd.get_dummies(df['Loan_Status'],drop_first=True)
df.drop(['Self_Employed',"Loan_Status"],axis=1,inplace=True)
df = pd.concat([df,self_employed,loan_status],axis=1)
```

```python
df.rename(columns={"Yes":"Self_Employed"} ,inplace=True)
df.rename(columns={"Y":"Loan_Status"},inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, stratify = y)
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

### Voting

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
model1 = LogisticRegression()
model2 = DecisionTreeClassifier()
model3 = SVC()
```

```python
model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model3.fit(X_train, y_train);
```

```python
model1_pred = model1.predict(X_test)
model2_pred = model2.predict(X_test)
model3_pred = model3.predict(X_test)
```

```python
model1_pred
```

```python
model3_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(model1.__class__.__name__ , accuracy_score(y_test, model1_pred))
print(model2.__class__.__name__ , accuracy_score(y_test, model2_pred))
print(model3.__class__.__name__ , accuracy_score(y_test, model3_pred))
```

### HARD

```python
from sklearn.ensemble import VotingClassifier
```

```python
estimators = [('lr', model1), ('dt', model2) , ('svm', model3) ] 
voting_hard = VotingClassifier(estimators, voting="hard")
```

```python
voting_hard.fit(X_train, y_train)
```

```python
voting_hard_pred = voting_hard.predict(X_test)
```

```python
accuracy_score(y_test, voting_hard_pred)
```
